```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import pandas_datareader
import datetime
from pandas_datareader import data as web
from quandl_key import quandl_api
```

```python
#The github version of this will not work without an api key for the quandl api being inserted into variable 'quandl_api'. The local version of this has my api key saved in an alternate file.

start = datetime.datetime(2012, 1 ,1)
end = datetime.datetime(2018, 3, 27)
delta = web.DataReader("DAL", 'quandl', start, end, api_key=quandl_api)

delta.head()
```

```python
JB = web.DataReader("JBLU", 'quandl', start, end, api_key=quandl_api)
SW = web.DataReader("LUV", 'quandl', start, end, api_key=quandl_api)
```

```python
JB.head()
```

```python
SW.head()
```

```python
delta['AdjClose'].plot(label='DAL', figsize=[16,8], title="Adjusted Closing Price")
JB['AdjClose'].plot(label='JBLU')
SW['AdjClose'].plot(label='LUV')
plt.legend()
```

```python
delta['Volume'].plot(label='Dal', figsize=(16, 8), title="Volume Traded")
JB['Volume'].plot(label='JBLU')
SW['Volume'].plot(label='LUV')
plt.legend()
```

```python
delta['Volume'].idxmax()
```

```python
delta['MA50'] = delta['AdjClose'].rolling(50).mean()
delta['MA200'] = delta['AdjClose'].rolling(200).mean()
delta[['AdjClose', 'MA50', 'MA200']].plot(label='DAL', figsize=(16, 8), title='Delta Adjusted Close and 50 and 200 Day Moving Averages')
```

```python
JB['MA50'] = JB['AdjClose'].rolling(50).mean()
JB['MA200'] = JB['AdjClose'].rolling(200).mean()
JB[['AdjClose', 'MA50', 'MA200']].plot(label='DAL', figsize=(16, 8), title='Jet Blue Adjusted Close and 50 and 200 Day Moving Averages')
```

```python
SW['MA50'] = SW['AdjClose'].rolling(50).mean()
SW['MA200'] = SW['AdjClose'].rolling(200).mean()
SW[['AdjClose', 'MA50', 'MA200']].plot(label='DAL', figsize=(16, 8), title='Southwest Adjusted Close and 50 and 200 Day Moving Averages')
```
